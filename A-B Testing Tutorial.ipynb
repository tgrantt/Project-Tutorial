# Introduction/Motivation

A/B Testing is quite simple. 

Imagine youre back in elementary school learning about the scientific method. You ask a question, you do your research, you create your hypothesis, then you test your hypothesis with an experiment and then you analyze your results and determine your conclusions. 

A/B testing is basically that. A/B testing is when one wants to know if a certain feature (test variable) has a certain effect (measured metric). You generate a null hypothesis and an alternate hypothesis and then you create the A/B test. Group A can be the control group with nothing changed and group B can have your new feature or whatever it is youre changing but everything else must be consistent. The two groups need to be as identical as possible except for this one variable change. Then data is generated and then the results are taken and statistical hypothesis testing is done to determine if that one changed feature had a statistically significant change on the targeted metric. Once that is determined a company can move forward with the next test which is formatted the same.

Many companies use this form of testing to validate different aspects of their product. Often times companies will compare different aspects of their website or online product to see if different features lead to a better result. Its easy to do A/B testing with online products like websites because you can just change a feature and measure its impact instantly while changing a feature on a physical product is a bit harder and a bit harder to measure its impact. For example Netflix adding a feature where you can become "friends" with other netflix users and them testing to see if this increases subscriptions or not is an example of a change that can be launched and its impact can be measured within days. If nike changed something on one of their shoes it would be dificult to have a control group and a test group and measure a single metric on those two groups.

# Tutorial


For this tutorial we will imagine we are at Netflix and we want to test a new feature on the netflix platform. This new feature will be a new way to display featuerd content on the home page when a user enters netflix. When they first open it there will be two highlighted movies instead of one with play buttons for each. We will test to see how this feature effects the number of clicks before the user watches something.

Hypothesis:
    
    Group A
        With No New Feature there will be More clicks to view

    Group B
        With a New feature there will be Fewer clicks to view
    
Metric to measure:
    
    Number of clicks
    
Now below is a generated data set with potential results from the A/B test.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.DataFrame()
```

```python
# We will test across 100,000 users 
df['user_id'] = np.array([num for num in range(100000)])

# 50% will be in the control group and 50% will be in the test group
df.loc[df['user_id'] < 50000,'group'] = 'control-A'
df.loc[df['user_id'] >= 50000,'group'] = 'test-B'

# for our dataset we will generate random number of clicks using the poisson distribution since it is continuous
df.loc[df.group=='test-B', 'num_clicks'] = np.random.poisson(4.9, 50000)
df.loc[df.group=='control-A', 'num_clicks'] = np.random.poisson(4.4, 50000)
```

```python
# verify our split
print('Test subjects:\t\t',len(df[df['group']=='test-B']))
print('Control subjects:\t',len(df[df['group']=='control-A']))
```

```python
# just get an idea for the different number of clicks in our data
df.num_clicks.sort_values().unique()
```

```python
# view the ditribution of the number of clicks by group using pyplot
plt.subplot(1,2,1)
plt.hist(df[df.group=='control-A'].num_clicks,color='green')
plt.subplot(1,2,2)
plt.hist(df[df.group=='test-B'].num_clicks,color='red')
plt.tight_layout()
```

```python
# view the ditribution of the number of clicks by group using seaborn
plt.subplot(1,2,1)
sns.distplot(df[df.group=='control-A'].num_clicks,color='green')
plt.subplot(1,2,2)
sns.distplot(df[df.group=='test-B'].num_clicks,color='red')
plt.tight_layout()
```

```python
# print the p-value to see if the results are statistically significant and not just from random chance.
print('P-Value: ',(stats.ttest_ind(df[df.group=='control-A'].num_clicks,df[df.group=='test-B'].num_clicks))[1])
```

**Results**

Since our p-value is lower than industry standard of 5% then we reject our null hypothesis and our can conclude that our results are statistically significant.

**Things to pay attention to:**

Pay attention to your data set and each of the aspects of your users. You will want to randomly sample your data across all aspects of the data. For example what if your new feature had a strong effect on people over the age of 65 and a negative effect on people under 65. And what if your test group consisted of 75% of the 65+ year old people. Then your results would be significant but when you go to put the new feature to the entire group of users your results there will be skewed and not as significant as your A/B test predicted. 

The way to combat this issue is random sampling. Ensure that when applying the tests and splitting the groups into your test and control group that each have roughly the same percentage of people with the same demographics and habits to ensure the best results. Also we want to test for as long as we can. The longer we test, with the most users we can test the higher the statistical power will be and the better you can combat **type I** and **type II** errors. 

In addition there are times to use a ONE-WAY ANOVA test or a TWO-WAY ANOVA. For One-way if you're just looking at users that are 65+ years old. In this form of testing it is important to always pay attention to the aspects of the users you are testing to ensure it is an equal playing field for your group A and group B. For A/B testing you want everything across the board to be equal so that your results are the most accurate and consistent. The only thing you want to be different between the groups is that one feature. That way you know that the only thing causing a change in your metric to measure is that one feature. 

**Time-based issues** can occur. For example, changing the color of the 'ADD TO CART' button could show a statistically significant change in user interaction with the page but this could be strictly because it was something they didnt pay attention to but now that it is a new color they noticed it. Over time however, they will stop paying attention to it again which will show no significance. That is why it is important to run the test long enough to account for time-based issues.
